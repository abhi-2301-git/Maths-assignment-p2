Based on a random sample drawn from the dataset, how does the sample mean of a key numerical variable compare to the population mean? 
What might explain any differences?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
df = pd.read_csv(r'E:\Code\Maths assignment\tested.csv')
```

```python
# grouping data by first class tickets
newdf = df[df['Pclass'] == 1][['Fare']]
```

```python
sns.histplot(data=newdf, x='Fare', kde=True)
plt.title('KDE Plot of Fare')
plt.show()
```

```python
# Since very few(1) have paid a fare above 300, it's best to remove that value as it could drastically affect the mean.
newdf = newdf[newdf['Fare'] < 300]
sns.histplot(data=newdf, x='Fare', kde=True)
plt.title('KDE Plot of Fare')
plt.show()
```

```python
# Now we can find population mean and sample mean and compare them both
MeanFare = np.mean(newdf['Fare'])
MeanFare
```

```python
# taking 3 sample out
samp1 = newdf['Fare'].sample(n=25, random_state=42)
samp2 = newdf['Fare'].sample(n=25, random_state=43)
samp3 = newdf['Fare'].sample(n=25, random_state=44)
```

```python
# calculating sampling mean
Smean1=samp1.mean()
Smean2=samp2.mean()
Smean3=samp3.mean()
```

```python
print(f'Mean 1: {Smean1}')
print(f'Mean 2: {Smean2}')
print(f'Mean 3: {Smean3}')
FirstClassMeanFare= newdf['Fare'].mean()
print(f'Pop mean-{FirstClassMeanFare}')
```

Since we can see Smean1 is the closest to population mean and the other two are +6.59523237735847 and 
 -18.17109962264152 away form population mean 
 this could be due to sampling variability, as each sample consists of different randomly chosen values
