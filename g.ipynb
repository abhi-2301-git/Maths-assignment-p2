```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'E:\Code\Maths assignment\tested.csv')

df = df[df['Fare'] <= 300]
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# Prepare the data
X = df[['Pclass', 'Age', 'SibSp', 'Parch']].fillna(0)  # Features
y = df['Fare'].fillna(0)  # Target

# Fit the model
model.fit(X, y)
```

```python
# Predict the target variable using the model
y_pred = model.predict(X)

# Plot the actual vs predicted values
plt.scatter(y, y_pred, alpha=0.5, label='Predicted vs Actual')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--', label='Ideal Fit')
plt.xlabel('Actual Fare')
plt.ylabel('Predicted Fare')
plt.title('Actual vs Predicted')
plt.legend()
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

Intercept (99.85):
When all features (Pclass, Age, SibSp, and Parch) are zero, the predicted fare is $99.85. (This is just a baseline — not realistic in this context.)

Pclass (-34.48):
For each increase in passenger class (e.g., from 1st to 2nd), the fare is expected to decrease by $34.48, assuming all other variables are constant. This makes sense because higher class numbers mean lower travel class (1st class = best, 3rd class = lowest).

Age (0.21):
Each additional year of age slightly increases the fare by $0.21, holding other variables constant.

SibSp (7.65):
Each additional sibling/spouse aboard increases the fare by $7.65, likely because they're booked together.

Parch (10.98):
Each additional parent/child aboard increases the fare by $10.98, also suggesting group travel raises the total fare.
